```python
bag=pd.read_csv('D:/Words.csv')
d = {}
for a, x in bag.values:
    d[a] = x


wordcloud = WordCloud()
wordcloud.generate_from_frequencies(frequencies=d)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
